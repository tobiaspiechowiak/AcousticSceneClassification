```python
#import necessary packages
import pandas as pd
import numpy as np
from sklearn.neural_network import MLPClassifier 
import soundfile as sf
```

```python
#import data 
path = 'x:/Steering/Recordings/Processed/'
data, fs = sf.read(path + 'ERB=1.wav')
classes = pd.read_csv(path + 'Classes_1.csv')
#data = pd.read_csv(path + 'Data_EBR=1.csv')
```

```python
X = data[:len(data)-1,:215]
Y = classes.values[:,0]
```

```python
len(X)
```

```python
#split into training and test set
from sklearn.model_selection import train_test_split
```

```python
trainX, testX, trainY, testY = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
#do one-hot encoding
from sklearn.preprocessing import OneHotEncoder 
enc = OneHotEncoder(sparse=False)
#for the train set
integer_encoded_train = trainY.reshape(len(trainY), 1)
onehot_encoded_train = enc.fit_transform(integer_encoded_train)
#for the test set
integer_encoded_test = testY.reshape(len(testY), 1)
onehot_encoded_test = enc.fit_transform(integer_encoded_test)
```

```python
onehot_encoded_train, onehot_encoded_test
```

```python
#define neural net  216-36-4 
clf = MLPClassifier(hidden_layer_sizes=(36,),solver='adam',
                    activation='tanh',max_iter=500,verbose = True,
                    random_state=0,tol = 1e-4).fit(trainX,onehot_encoded_train)
```

```python
clf.score(trainX,onehot_encoded_train)
```

```python
clf.score(testX,onehot_encoded_test)
```
